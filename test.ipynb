```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
import matplotlib.pyplot as plt
import time
import csv
```

```python
#Set up Splinter
browser = Browser("chrome")
#Set up base url

url = "https://projects.propublica.org/nonprofits/organizations/850403577"
```

```python
browser.visit(url)
```

```python
#Parse the HTML
html = browser.html

#Create a BeautfulSoup object from the scrapped html

market_soup = soup(html, "html.parser")

#Check if HTML was scrapped correctly
```

```python
market_soup
```

```python
revenue_div = market_soup.find_all("div", class_ ='row-revenue__number stack__item--size-1')
revenue_list = [rev.text.strip() for rev in revenue_div]

expenses_div = market_soup.find_all("div", class_ ='row-summary__number')
expenses_list = [exp.text.strip() for exp in expenses_div]

revenue_list
#expenses_list

expenses_final_list  = expenses_list[0::3]
expenses_final_list

```

```python
year =[2022, 2021, 2020,2019,2018,2017,2016,2015,2014,2013,2012,2011]

combined_data = zip(year,revenue_list,expenses_final_list)

with open ("non-profit.csv", 'w',newline ='') as file :
    writer = csv.writer(file)
    writer.writerow(["Year","Revenue","Expenses"])
    writer.writerows(combined_data)
```
